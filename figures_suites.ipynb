```python
import pylab as pl
import numpy as np

import matplotlib as mpl
import matplotlib.font_manager as font_manager

mpl.rcParams['font.family'] = 'serif'
cmfont = font_manager.FontProperties(fname=mpl.get_data_path() + '/fonts/ttf/cmr10.ttf')
#mpl.rcParams['font.serif']=cmfont.get_name()
mpl.rcParams['mathtext.fontset']='cm'
mpl.rcParams['axes.unicode_minus']=False
```

```python
fig, ax = pl.subplots(figsize=(10, 8))
        
        
##### Axes modifi√©s
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',6))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

ax.axis([-0.1, 10.1, 6, 27.2])
        

##### Graduations ordonn√©es
ax.set_xticks(range(11))
ax.set_xticklabels([r'$\mathbf{'+str(k)+'}$' for k in range(11)],
                                color = 'black',fontsize = 15)
        
ax.set_yticks(range(7, 29, 2))
ax.set_yticklabels([r'$\mathbf{'+str(k)+'}$' for k in range(7, 29, 2)],
                                color = 'black',fontsize = 15)
        
        
ax.tick_params( direction = 'inout',width = 2, size = 12, 
               colors = 'black', labelsize = 16)

ax.grid(True, which="major", ls="-", color='black', linewidth = 0.2)
ax.grid(True, which="minor", ls="-", color='gray', linewidth = 0.3)

hfont = {'fontname' : 'Comic Sans MS', 'weight' : 'bold', 'size' : 18}
hfont1 = {'fontname' : 'Comic Sans MS', 'weight' : 'bold', 'size' : 22, 
          'color' : 'blue'}

boite = dict(facecolor='lightgray', edgecolor='black', boxstyle='round,pad=0.2')

X = [k for k in range(11)]
u = 7
Y = [u]
for k in range(1, 11):
    u = u + 2
    Y.append(u)

for k in range(10):
    if k == 4:
        ax.plot([X[k], X[k+1], X[k+1]], [Y[k], Y[k], Y[k+1]],
                '-', color = 'red', linewidth=3)
    else:
        ax.plot([X[k], X[k+1], X[k+1]], [Y[k], Y[k], Y[k+1]],
                '--', color = 'gray', linewidth=2)
    
    
for k in range(11):
    ax.text(X[k], Y[k]+0.3, r'$P_{'+str(k)+'}$', **hfont1, 
            ha = 'right')

ax.plot(X, Y, 'o-', color = 'blue',
        markerfacecolor = 'white', markersize = 9)

ax.text((X[4] + X[5])/2, Y[4], 
        r'$\mathbf{+1}$', **hfont, 
        bbox = boite, ha = 'center', va = 'top')
ax.text(X[5], (Y[4] + Y[5])/2, 
        r'$\mathbf{+2}$', **hfont, 
        bbox = boite, ha = 'left', va = 'center')
ax.text(X[5]+0.5, (Y[4] + Y[5])/2, 
        r'$=$ raison', **hfont, 
        ha = 'left', va = 'center')


ax.set_title(r'Repr√©sentation de $(u_n)$ o√π $u_{n+1}=u_n+2$ avec $u_0=7$',**hfont,bbox=boite)

fig.tight_layout()
fig.savefig('suite_arithmetique.eps')
```

```python
fig, ax = pl.subplots(figsize=(10, 8))
        
        
##### Axes modifi√©s
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)

ax.axis([-0.1, 10.1, 0, 7.1])
        

##### Graduations ordonn√©es
ax.set_xticks(range(11))
ax.set_xticklabels([r'$\mathbf{'+str(k)+'}$' for k in range(11)],
                                color = 'black',fontsize = 15)
        
ax.set_yticks(range(8))
ax.set_yticklabels([r'$\mathbf{'+str(k)+'}$' for k in range(8)],
                   color = 'black',fontsize = 15)
        
        
ax.tick_params( direction = 'inout',width = 2, size = 12, 
               colors = 'black', labelsize = 16)

ax.grid(which="major", ls="-", color='black', linewidth = 0.2)
ax.grid(which="minor", ls="-", color='gray', linewidth = 0.3)

hfont = {'fontname' : 'Comic Sans MS', 'weight' : 'bold', 'size' : 18}
hfont1 = {'fontname' : 'Comic Sans MS', 'weight' : 'bold', 'size' : 22, 
          'color' : 'blue'}

boite = dict(facecolor='lightgray', edgecolor='black', boxstyle='round,pad=0.2')

X = [k for k in range(11)]
u = 1
Y = [u]
for k in range(1, 11):
    u = 1.2 * u
    Y.append(u)

for k in range(10):
    if k == 4:
        ax.plot([X[k], X[k+1], X[k+1]], [Y[k], Y[k], Y[k+1]],
                '-', color = 'red', linewidth=3)
    else:
        ax.plot([X[k], X[k+1], X[k+1]], [Y[k], Y[k], Y[k+1]],
                '--', color = 'gray', linewidth=2)
    
    
for k in range(11):
    ax.text(X[k], Y[k]+0.1, r'$P_{'+str(k)+'}$', **hfont1, 
            ha = 'right')

ax.plot(X, Y, 'o-', color = 'blue',
        markerfacecolor = 'white', markersize = 9)

ax.text((X[4] + X[5])/2, Y[4], 
        r'$\mathbf{+1}$', **hfont, 
        bbox = boite, ha = 'center', va = 'top')
ax.text(X[5]+0.1, (Y[4] + Y[5])/2, 
        r'$\mathbf{\times 1,2}$', **hfont, 
        bbox = boite, ha = 'left', va = 'center')
ax.text(X[5]+0.7, (Y[4] + Y[5])/2, 
        r'$=$ raison', **hfont, 
        ha = 'left', va = 'center')

ax.set_title(r'Repr√©sentation de $(u_n)$ o√π $u_{n+1}=1,2\times u_n$ avec $u_0=1$',**hfont,bbox=boite)



fig.tight_layout()
fig.savefig('suite_geometrique.eps')
```
