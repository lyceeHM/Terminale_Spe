## Question n°1

```python
import pylab as pl
from math import *

t = 0
y = 0.01
X = [t]
Y = [y]
for n in range(1, 41):
    y = y + 0.4 * y * (1 - y)
    t = t + 1
    X.append(t)
    Y.append(y)
    
pl.plot(X, Y, 'o-')
pl.show()
```

## Question n°2

```python
Sol = [1 / (1 + 99 * exp(- 0.4 * x)) for x in X]

pl.plot(X, Sol, color = 'red', label = 'solution exacte')
pl.plot(X, Y, 'o-', label = 'Euler')
pl.legend()
pl.show()
```
