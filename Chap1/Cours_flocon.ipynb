Exécuter les instructions suivantes pour visualiser les étapes de constructions du flocon de Von Koch.
</blockquote>

```python
from turtle import *
```

```python
couleur = ['red', 'black'] * 2

def von(n, d):
    if n == 0:
        fd(d)
    else:
        von(n - 1, d / 3)
        lt(60)
        von(n - 1, d / 3)
        rt(120)
        von(n - 1, d / 3)
        lt(60)
        von(n - 1, d / 3)

setworldcoordinates(-15,-100,116,32)
#setworldcoordinates(-1,-100,70,50)
speed("fast")
def etapes(N):
    t = Turtle()
    t.hideturtle()
    for n in range(1, N + 1):
        reset()
        color(couleur[n - 1])
        begin_fill()
        width(N + 1 - n)
        von(n, 100)
        rt(120)
        von(n, 100)
        rt(120)
        von(n, 100)
        end_fill()
        
        # Affichage du texte
        for i in range(1, n + 1):
            t.up()
            t.setpos(50, -30)
            t.pd()
            t.color(couleur[n - 1])
            t.write('Etape '+str(i - 1),align = 'center', 
                    font = ('comic sans ms', 40, 'normal'))
        t.color('white')
        t.write('Etape '+str(n),align = 'center', 
              font = ('comic sans ms', 40, 'normal'))
        
        
etapes(3)
hideturtle()
done()
```

<blockquote style="border-left: 3px solid black">
    <b>On démontre que : </b>
    <ol>
        <li>&nbsp;&nbsp; $c_n=3\times 4^n$ &nbsp;&nbsp;&nbsp;&nbsp;<i>(par récurrence)</i></li>
        <li>&nbsp;&nbsp; $\ell_n=\dfrac{1}{3^n}$ &nbsp;&nbsp;&nbsp;&nbsp;<i>(par récurrence)</i></li>
        <li>&nbsp;&nbsp; $p_n=3\times\left(\dfrac{4}{3}\right)^n$</li>
        <li>&nbsp;&nbsp; $A_{n+1}=A_n + \dfrac{\sqrt{3}}{12}\times \left(\dfrac{4}{9}\right)^n$ </li>
        <li>&nbsp;&nbsp; $A_n=\dfrac{\sqrt{3}}{4}+\dfrac{3\sqrt{3}}{20}\left(1-\left(\dfrac{4}{9}\right)^n\right)$ &nbsp;&nbsp;&nbsp;&nbsp;<i>(par récurrence)</i></li>
    </ol>
</blockquote>    

```python
from math import sqrt

p = 3
A = sqrt(3) / 4
print('{0: <10f} | {1: <10f}'.format(p, A))
for n in range(1, 20):
    p = p * 4 / 3
    A = A + sqrt(3) / 12 * (4 / 9) ** (n - 1)
    print('{0: <10f} | {1: <10f}'.format(p, A))
```
