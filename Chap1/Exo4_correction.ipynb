```python
u = 6
for k in range(1, 21):
    u = (3 * u - 2) ** 0.5
    print('{0: <3} | {1: <20}'.format(k, u))
```

<blockquote style="border-left: 3px solid black"> On retrouve que ${\displaystyle \lim_{n\rightarrow+\infty}u_{n}=2}$.
</blockquote>

<span style="font-size:2em;font-weight:bold">Bonus.</span>&nbsp;

```python
import pylab as pl
pl.style.use('bmh')

def f(x):
    return (3 * x - 2) ** 0.5

u = 6
X = [u, u]
Y = [0, f(u)]

fig, ax = pl.subplots(figsize=(8, 8))
ax.plot(X, Y, '--',color = 'black')

X = [u]
Y = [f(u)]
for k in range(1, 21):
    u = f(u)
    X = X + [u, u]
    Y = Y + [u, f(u)]


        
##### Axes modifiés
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
    
ax.tick_params(direction = 'inout',width = 1, size = 9, 
               colors = 'black', labelsize = 13)   
A = [2 / 3 + k / 100 for k in range(600)]    
ax.plot(A, [f(a) for a in A], color = 'red')
ax.plot([0, 6], [0, 6], color = 'green')
ax.plot(X, Y, color = 'black')

ax.plot(2, 2, 'o', markerfacecolor = "white", markeredgecolor = 'red')
```

<blockquote style="border-left: 3px solid black"> Il y avait deux points fixe de $\left(u_n\right)$ : $1$ et $2$. <br> C'est vers le le premier que $\left(u_n\right)$ convergea.
</blockquote>
