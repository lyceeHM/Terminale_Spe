```python
import pandas as pa
import pylab as pl

T = pa.read_csv('crop_circle.csv')
T.head()
```

```python
T[[x for x in T.columns if x not in ['Year', 'Month', 'Country']]].mean().plot.pie()
```

```python
P = T.groupby('Country')[['wheat', 'barley', 'other']].mean()
P.T
```

```python
T['Country'].value_counts()
```

```python
fig, ax = pl.subplots(figsize = (9, 4))
P.T.plot.pie(
    ax = ax,
    subplots = True, 
    layout = (3, 3),
    labels = None,
    autopct = lambda x : str(int(x)) + '%' if x > 0 else '',
    legend = None,
    radius = 1,
    pctdistance = 1.2,
    colors = ['lightgray', 'white', 'pink'],
    wedgeprops = {'edgecolor' : 'black'}
)

fig.legend(bbox_to_anchor = (1, 0.6), 
          labels = ['bl', 'orge', 'autre'],
          ncol = 1,
          title = 'Type')
fig.tight_layout()
pl.show()
```
